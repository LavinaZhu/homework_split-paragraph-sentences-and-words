```python
#分句 
from urllib import request
import nltk, re, pprint
from nltk import word_tokenize
import jieba
raw = open("F:\chinese_eg.txt","r",encoding="UTF-8")

cutlist ="。！？"  
def FindToken(cutlist, char):  
    if char in cutlist:  
        return True  
    else:  
        return False      
def Cutsentence(cutlist,lines):           
    l = []          
    line = []      
          
    for i in lines:         
        if FindToken(cutlist,i):       
            line.append(i)         
            l.append(''.join(line))    
            line = []   
        else:        
            line.append(i)       
    return l  

for lines in raw:      
    l = Cutsentence(list(cutlist),list(lines))
    for line in l:    
        if line.strip() !="":        
            li = line.strip().split()
            for sentence in li:  
                print (sentence)
```

```python
#结巴分词
from urllib import request
import nltk, re, pprint
from nltk import word_tokenize
import jieba
raw = open("F:\chinese_eg.txt","r",encoding="UTF-8")

def Cutwords():
    for lines in raw:      
        l = Cutsentence(list(cutlist),list(lines))
        for line in l:    
            if line.strip() !="":        
                li = line.strip().split()
                for sentence in li:
                    seglist=jieba.cut(sentence)
                for seg in seglist:
                    words=''
                    words+=seg
                    print(words)
if  __name__ == '__main__':  
    Cutwords()
```
